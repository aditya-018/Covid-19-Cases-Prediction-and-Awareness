```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn import datasets
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import GridSearchCV
```

```python
df_train = pd.read_csv("training_data.csv")
df_train
```

```python
df_test = pd.read_csv("test_data.csv")
df_test
```

```python
X_train = pd.concat([df_train.iloc[:,2], df_train.iloc[:, 4:]], axis=1)
Y_train = df_train.iloc[:,1].values
X_train
```

```python
Y_train
```

```python
X_test = pd.concat([df_test.iloc[:,2], df_test.iloc[:, 4:]], axis=1)
X_test
```

```python
selector = SelectKBest(score_func=f_regression, k=10)
X_train_selected = selector.fit_transform(X_train, Y_train)
X_test_selected = selector.transform(X_test)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
model = RandomForestRegressor(random_state=35)
grid_search = GridSearchCV(model, param_grid, cv=5, n_jobs=1)
grid_search.fit(X_train_selected, Y_train)
best_model = grid_search.best_estimator_
Y_pred = best_model.predict(X_test_selected)
Y_pred
```

```python
Y_pred = pd.DataFrame(Y_pred, columns= ['cases'])
```

```python
Y_pred
```

```python
#read the sample_submission file
sample_submission = pd.read_csv('sample_submission.csv')
sample_submission.head()
```

```python
#replace Cases values with the test_df_new['cases'] values
sample_submission['Cases'] = Y_pred
sample_submission['Cases'] = sample_submission['Cases'].round(0)
sample_submission['Cases'] = sample_submission['Cases'].mask(sample_submission['Cases'] < 0, 0)
sample_submission.head()
```
